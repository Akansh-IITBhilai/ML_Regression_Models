```python
def style_raw_data(styler):
    styler.background_gradient(
        subset=['MedHouseVal'],
        cmap='YlOrRd',
        vmax=df['MedHouseVal'].max(),
        vmin=df['MedHouseVal'].min()
    )
    styler.background_gradient(
        subset=['MedInc'],
        cmap='Blues',
        vmax=df['MedInc'].max(),
        vmin=df['MedInc'].min()
    )
    styler.format(precision=2)
    styler.set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#0077c2'), ('color', '#ffffff'), ('font-weight', 'bold')]},
        {'selector': 'td', 'props': [('border', '1px solid #e0e0e0')]},
        {'selector': 'table', 'props': [('border-collapse', 'collapse'), ('width', '100%')]}
    ])
    styler.set_caption("Selected California Housing Features (Raw Data)")
    return styler
```

```python
import pandas as pd
import numpy as np 
def mat(styler):
    idx = pd.IndexSlice
    styler.format(precision=2) \
        .highlight_min(subset=idx[['MSE', 'RMSE'], :], axis=1, color='#FFECB3') \
        .highlight_max(subset=idx['R2_Score', :], axis=1, color='#90EE90') \
        .set_table_styles([
            {'selector': 'th:not(.index)', 'props': [('background-color', '#4B5563'), ('color', 'white'), ('font-weight', 'bold')]},
            {'selector': '.index', 'props': [('background-color', '#F3F4F6'), ('color', '#1F2937'), ('font-weight', 'bold')]},
            {'selector': 'td', 'props': [('border', '1px solid #D1D5DB')]},
            {'selector': 'table', 'props': [('border-collapse', 'collapse'), ('width', '100%')]}
        ]) \
        .set_caption("Model Performance Evaluation Matrix (Best Scores Highlighted)")
    return styler
```
